#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('assets/real_estate.csv', sep=';')
df
cinturon_sur = df[df['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]
precio_max = cinturon_sur['price'].max()
precio_min = cinturon_sur['price'].min()

cinturon_sur['precios normalizados'] = (cinturon_sur['price'] - precio_min) / (precio_max - precio_min)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

ax1 = axes[0, 0]
fuenlabrada = cinturon_sur[cinturon_sur['level5'] == 'Fuenlabrada']
ax1.hist(fuenlabrada['precios normalizados'], bins=20, alpha=0.7)
ax1.set_title('Histograma de Precios Normalizados - Fuenlabrada')
ax1.set_xlabel('Precios Normalizados')
ax1.set_ylabel('Frecuencia')

ax2 = axes[0, 1]
leganes = cinturon_sur[cinturon_sur['level5'] == 'Leganés']
ax2.hist(leganes['precios normalizados'], bins=20, alpha=0.7)
ax2.set_title('Histograma de Precios Normalizados - Leganés')
ax2.set_xlabel('Precios Normalizados')
ax2.set_ylabel('Frecuencia')

ax3 = axes[1, 0]
getafe = cinturon_sur[cinturon_sur['level5'] == 'Getafe']
ax3.hist(getafe['precios normalizados'], bins=20, alpha=0.7)
ax3.set_title('Histograma de Precios Normalizados - Getafe')
ax3.set_xlabel('Precios Normalizados')
ax3.set_ylabel('Frecuencia')

ax4 = axes[1, 1]
alcorcon = cinturon_sur[cinturon_sur['level5'] == 'Alcorcón']
ax4.hist(alcorcon['precios normalizados'], bins=20, alpha=0.7)
ax4.set_title('Histograma de Precios Normalizados - Alcorcón')
ax4.set_xlabel('Precios Normalizados')
ax4.set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

Observamos que la mayoria de viviendas caen en el rango de precios baratos, entre 0,1 y 0,4, siendo muy pocas las que superan el 0,45, lo que significa que podemos encontrar pocas viviendas caras en estas poblaciones.

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('assets/real_estate.csv', sep=';')
df
cinturon_sur = df[df['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]
def calcular_valor(columna):
    return columna['price'] % columna ['surface']
cinturon_sur['pps'] = cinturon_sur.apply(calcular_valor, axis = 1)
plt.figure(figsize = (10, 5))

colores = {'Fuenlabrada': 'blue', 'Alcorcón': 'green', 'Leganés': 'red', 'Getafe': 'orange'}
etiquetas = {'Fuenlabrada': 'Fuenlabrada', 'Alcorcón': 'Alcorcón', 'Leganés': 'Leganés', 'Getafe': 'Getafe'}

for poblacion, color in colores.items():
    poblacion_data = cinturon_sur[cinturon_sur['level5'] == poblacion]
    plt.scatter(poblacion_data['price'], poblacion_data['pps'], 
                color=color, alpha=1, label=etiquetas[poblacion])
    
plt.title('Relación entre Precio y Precio por Metro Cuadrado')
plt.xlabel('Precio por metro cuadrado')
plt.ylabel('Precio')
plt.legend(title='Población')
plt.tight_layout()
plt.show()
```
